```python
mp_API_Key = '64aQ6cjeu5e7gMjzCJT'
mp_cryst = 'mp-11869'
```

```python
from pymatgen.ext.matproj import MPRester
import pymatgen.core.lattice as pcl



with MPRester(mp_API_Key) as m:
    structure = m.get_structure_by_material_id(mp_cryst)
    structure.to(filename = f'{mp_cryst}.cif')
    
    
    ##Voir : https://github.com/materialsproject/mapidoc/tree/master/materials
    #Les données qu'on recherche sont dans les dossiers "spacegroup" et "initial_struture"
    #On les extrait à l'aide de la fonction query 
    spaceGroup = m.query(criteria={"task_id": mp_cryst}, properties= ["spacegroup"])
    directStruct_lattice = m.query(criteria={"task_id": mp_cryst}, properties=["initial_structure.lattice"])
    
    #on stocke la matrice stockant les vecteurs de base du réseau direct 
    matrix = directStruct_lattice[0]['initial_structure.lattice']['matrix']

    #on crée un objet Lattice contenant toutes les données néc
    directLattice = pcl.Lattice(matrix)
    reciprocalLattice = directLattice.reciprocal_lattice

    directParams = directLattice.parameters
    reciprocalParams = reciprocalLattice.parameters


print("Direct Lattice Parameters :\n")
print("(a, b, c) = " + str(directParams[0:3]))
print("(alpha, beta, gamma = " + str(directParams[3:6]) + '\n')

print("Reciprocal Lattice Parameters :\n")
print("(a, b, c) = " + str(directParams[0:3]))
print("(alpha, beta, gamma = " + str(directParams[3:6]) +'\n')

print("Crystal System = " + spaceGroup[0]["spacegroup"]["crystal_system"]+'\n')
Lattice_type = spaceGroup[0]["spacegroup"]["symbol"][0]

if Lattice_type == 'F':
    print("Type of Lattice : Face centered\n")
elif Lattice_type == 'P':
    print("Type of Lattice : Primitive\n ")
if Lattice_type == 'I':
    print("Type of Lattice : Body Centered\n ")
if Lattice_type == 'C':
    print("Type of Lattice : Face Centered\n ")

print("Point Group : " + spaceGroup[0]["spacegroup"]["point_group"])

    
```

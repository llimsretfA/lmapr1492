```python
from pymatgen.ext.matproj import MPRester
from jupyter_jsmol import JsmolView
from IPython.display import display
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.core import Lattice, Structure
```

```python
mp_API_Key = '64aQ6cjeu5e7gMjzCJT'
mp_cryst = 'mp-11869'

with MPRester(mp_API_Key) as m:
    structure = m.get_structure_by_material_id(mp_cryst)
    structure.to(filename = f'{mp_cryst}.cif')

    lattice_data = m.query(criteria={"task_id": mp_cryst}, properties=["initial_structure.lattice"])

    #on construit l'objet Lattice de notre cristal
    lattice = Lattice(lattice_data[0]['initial_structure.lattice']['matrix'])

    #coordonnées des différents constituants (https://materialsproject.org/materials/mp-11869/#)
    coords = [[0, 0, 0], [0.5, 0.5, 0.5], [0.25, 0.25, 0.25]]

   #on construit l'objet Structure de notre cristal
    struct = Structure(lattice, ["Hf", "Sn", "Pd"], coords)

    view2 = quick_view(struct, conventional = True)
    display(view2)
```

## Détermination des 3 premiers pics de Diffraction du cristal HfSnPd pour un rayonnement X incident Cu Kα

Le premier import sert à obtenir les informations du cristal à partir de son fichier .cif sur le site https://materialsproject.org/

Le deuxième nous sert à calculer et afficher la figure de diffraction.

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

On commence par récupérer les informations de notre cristal, tout sera stocké dans un objet Structure défini dans le module pymatgen.core.structure.

```python
mp_API_Key = '64aQ6cjeu5e7gMjzCJT'
mp_cryst = 'mp-11869'

with MPRester(mp_API_Key) as m:
    structure = m.get_structure_by_material_id(mp_cryst)
```

On utilise ensuite la classe XRDCalculator qui va nous servir à calculer le paterne de diffraction et l'afficher pour notre structure à l'aide de la fonction get_plot() qui renvoi un objet matplotlib.pyplot auquel on applique la méthode show() pour afficher la figure de diffraction.  

On a défini le range de two_theta manuellement de façon à n'obtenir que les 3 premiers pics.

On obtient donc un premier pic en 2θ = 24.225° pour les indices (1,1,1) et (1,0,0), un deuxième pic en 2θ = 28.044° pour les indices (1,1,0) et un troisième pic en 2θ = 40.078° pour les indices (2,1,1) et (1,0,-1).

Les valeurs précises des angles à été déterminé par un appel à print(xrd.get_pattern(structure))

```python
xrd = XRDCalculator(wavelength='CuKa')

plot = xrd.get_plot(structure, two_theta_range=(0,45))

plot.show()
```
